However, we don't want to discredit the other neighborhoods. Let's look at these incidents on a map so we can see the crime distribution across Allegheny County.

```python
#first, let's create the base map
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") # read in the shapefile
```

Now that we've downloaded our shapefile, we're going to overlay our crime incident neighborhoods overtop this new map. 

```python
crime_map = neighborhoods.merge(crime, how='left', left_on='hood', right_on='INCIDENTNEIGHBORHOOD')
```

Now let's make the map a little prettier.

```python
crime_map.plot(column='INCIDENTNEIGHBORHOOD',
               cmap='BuPu',
               edgecolor="white",
               legend=True,
               figsize=(10, 10),
               missing_kwds={"color": "lightgrey"})
```
